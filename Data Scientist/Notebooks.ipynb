### Componentes del blucle FOR

- La palabra clave `for`.
- Nuestra variable de bucle `element`. En cada iteración del bucle toma el valor de un elemento de la secuencia y te permite realizar algunas operaciones en él. - Puedes usar cualquier nombre para esta variable, pero es mejor asegurarse de utilizar uno que sea claro y significativo.
- La palabra clave `in`.
- La lista `your_list`. El nombre de la lista que está siendo procesada por el bucle.
- Los dos puntos `:` son un elemento esencial que separa las dos partes del bucle.
- La función `print(element)`. Esto constituye el cuerpo del bucle. El cuerpo contiene comandos que el bucle ejecuta en cada iteración del bucle. Cada línea dentro del bucle debe estar indentada con 4 espacios, esto le indica a Python que esa acción se repite dentro del bucle, separado del resto del código.

También puedes utilizar bucles for con strings. Los bucles pueden iterar sobre cada carácter del texto, tal como lo hacen sobre cada elemento de la lista:

````python
review = 'The Emoji Movie dejó mucho que desear. Dos estrellas demasiado generosas.'
for word in review:
    print(word)
````

## Bucle WHILE

### Componentes del bucle WHILE
- La palabra clave while.
- La condición. Es una expresión que Python evalúa antes de cada iteración. Mientras la condición sea verdadera, el bucle seguirá ejecutándose.
- Los dos puntos :. Son un elemento esencial que separa la condición del cuerpo del bucle.
- La acción o cuerpo del bucle. Es el bloque de código que se ejecuta cada vez que la condición es verdadera. Cada línea dentro del cuerpo debe estar indentada con 4 espacios.
- La actualización o cambio de estado. Es una instrucción dentro del cuerpo del bucle que modifica algún valor relacionado con la condición. Esto permite que el bucle avance y evita que se repita infinitamente.

Si queremos calcular el peso de carga de un elevador, podemos hacer esto:

````python
from random import randint # Importa la función randint de la librería random
elevator_capacity = 500
total_weight = 0 # Inicializa el peso total
person_count = 0 # Inicializa el contador de personas

while total_weight < elevator_capacity: # Variable de control del bucle, se ejecuta mientras el peso total sea menor o igual a la capacidad del ascensor
    person_weight = randint(50, 110) # Genera un peso aleatorio entre 50 y 110 kg
    total_weight += person_weight # Suma el peso de la persona al peso total usando funciones de asignación ampliada
    person_count += 1 # Incrementa el contador de personas
    print(f'Peso actual: {total_weight}kg. Cantidad de personas: {person_count}')
print(f'El ascensor llegó a {total_weight}kg. Cantidad de personas: {person_count}')
````

Sin embargo, el bucle suma el peso de la persona sin comprobar primero si esa persona cabe sin exceder la capacidad del ascensor. Para resolverlo, hacemos esto:

````python
from random import randint # Importa la función randint de la librería random
elevator_capacity = 500
total_weight = 0 # Inicializa el peso total
person_weight = randint(50, 110) # Genera un peso aleatorio entre 50 y 110 kg
person_count = 0 # Inicializa el contador de personas

while total_weight + person_weight <= elevator_capacity: # Validamos si la persona cabe. En el código anterior agregábamos a la persona en la ejecución del bucle y por eso aquí la creamos antes del bucle.
    total_weight += person_weight # Suma el peso de la persona al peso total usando funciones de asignación ampliada
    person_count += 1 # Incrementa el contador de personas
    print(f'Peso actual: {total_weight}kg. Cantidad de personas: {person_count}')
print(f'El ascensor llegó a {total_weight}kg. Cantidad de personas: {person_count}')
````

Son palabras clave que se utilizan para comparar dos valores y devolver un resultado booleano.

| Operador | Significado | Ejemplo | Resultado |
| --- | --- | --- | --- |
| `==` | Igualdad | 5 == 5 | True |

- La condición, por ejemplo: `weather == "lluvia"`.

- La indentación del bloque `else`.
  - La acción alternativa también debe ir indentada con 4 espacios.

### Múltiples ramas con ELSE y ELIF
La sentencia `elif` es una abreviatura de `else if` y permite comprobar condiciones adicionales después de una sentencia `if` inicial.

El uso de la palabra clave `elif` nos permite agregar cualquier número de ramas. Sin embargo, solo se ejecutará una rama: aquella cuya condición se cumpla primero. Todas las demás ramas serán ignoradas.

Por eso es importante ordenar correctamente las ramas, con las condiciones más específicas primero y las más generales al final. Al hacer esto, podemos asegurarnos de que el código se comporte como esperamos y que se cubran todos los casos posibles.

````python
years = [1994, 1972, 2008, 1993, 2003, 1994, 1966, 1999, 1962, 1997, 0]

for year in years:
    if 1960 <= year <= 1969:
        print('La película se estrenó en la década de los 60\'s')
    elif 1970 <= year <= 1979:
        print('La película se estrenó en la década de los 70\'s')
    elif 1980 <= year <= 1989:
        print('La película se estrenó en la década de los 80\'s')
    elif 1990 <= year <= 1999:
        print('La película se estrenó en la década de los 90\'s')
    elif 2000 <= year <= 2009:
        print('La película se estrenó en la década de los 2000')
    else:
        print('No existe una fecha para la película')
````

### Modificación de tablas

Trabajaremos en una tabla con las siguientes columnas: `Name`, `Salary`, `Position` y `Seniority Level`. Aumentaremos el salario de un empleado en específico y, si el nuevo salario supera un determinado umbral especificado, actualizaremos su `Seniority Level`.

Por ejemplo, veremos lo siguiente: 

- `name_to_update` = "Alicia" (empleado a analizar)
- `salary_increase` = 10.000 (valor en el que se incrementará su salario)
- `salary_threshold` = 80.000 (límite para cambiar de rango, si el salario actualizado supera ese tope se deberá actualizar su rango)
`new_seniority` = "Senior" (nuevo rango a implementar si supera el tope salarial)

Entonces, si al salario actual de Alicia le sumo $10.000 y el resultado supera los $80.000, entonces su rango se deberá cambiar a Senior.

````python
employees = [['Alice', 75000, 'Desarrollador', 'Júnior'],
			['Bob', 68000, 'Diseñador', 'Semi sénior'],
			['Charlie', 78000, 'Desarrollador', 'Júnior'],
			['David', 85000, 'Gerente', 'Sénior'],
			['Eve', 80000, 'Desarrollador', 'Semi sénior']]


# Define el nombre que actualizarás, el incremento de salario y el umbral especificado de salario para el cambio de antigüedad
name_to_update = "Alice"
salary_increase = 10000
salary_threshold = 80000
new_seniority = "Sénior"

# Modifica la tabla
for employee in employees:
    if employee[0] == name_to_update:
        # Incrementa el salario
        employee[1] += salary_increase
        # Comprueba si el nuevo salario supera el umbral especificado
        if employee[1] > salary_threshold:
            # Actualiza el grado de antigüedad
            employee[3] = new_seniority

# Muestra la tabla actualizada
print(employees)
````

# Diccionarios
Cada entrada del diccionario de Python tiene dos partes: una clave y un valor.

- `Clave`: un string o número que identifica el valor.
- `Valor`: los datos asociados a la clave.

Estas son las tres reglas básicas de la sintaxis del diccionario:

- Los diccionarios se escriben con llaves: `{}`.
- Las comas separan los elementos (pares clave-valor) dentro del diccionario.
- Una `clave` y un `valor` dentro de un par se separan por dos puntos.


````python
# Diccionario anidado con los productos en inventario
products = {
    'Product_A': {'price': 10, 'quantity': 2, 'discount': 0.1},
    'Product_B': {'price': 20, 'quantity': 1, 'discount': 0.05},
    'Product_C': {'price': 15, 'quantity': 5, 'discount': 0.2},
}

# Paso 1: Inicializa la variable para almacenar el total
total_cost = 0

# Paso 2: Recorre cada producto
for product in products:
    # Accede al precio y a la cantidad
    price = products[product]['price']
    quantity = products[product]['quantity']

    # Paso 3: Calcula el ingreso por producto y acumúlalo
    ### TU CÓDIGO AQUÍ ###
    cost = price * quantity
    total_cost += cost

# Paso 4: Muestra el resultado final
print(f"Ingresos totales de la tienda: ${total_cost}")
````

````python
clients = [
    [32456, "Jack Wilson", 31, 150000, "Healthcare"],
    [34591, "Nina Brown", 43, 250000, "Telecom"],
    [37512, "John Smith", 32, 210000, "IT"],
    [39591, "Brian Perez", 49, 340000, "Healthcare"],
    [45123, "Brian Lee", 48, 75000, "Telecom"],
    [47635, "David Chen", 56, 180000, "Telecom"],
    [49571, "Brian Chen", 52, 220000, "IT"],
    [50391, "David Rodriguez", 31, 120000, "IT"],
    [34556, "Lucas Hernandez", 37, 180000, "Healthcare"]
]

incomes_per_field = {} # aquí colocarás los ingresos para cada campo

for client in clients:
	# primero, extrae el nombre del campo
    field = client[4]
	# segundo, extrae los ingresos
    income = client[3]

    if incomes_per_field.get(field) == None: # comprueba si el campo extraído NO está en el diccionario incomes_per_field
        incomes_per_field[field] = [income] # añade un nuevo campo como clave y establece una lista como valor 
    else: # si el campo extraído está en el diccionario incomes_per_field
        incomes_per_field[field].append(income) # añade el nuevo ingreso a la lista de ingresos para un campo en particular
		

# no modifiques el código de abajo, ya que imprime el resultado
print(incomes_per_field)
````

# Funciones
- Primero viene el encabezado de la función.
- La palabra clave `def` se utiliza para especificar el nombre de la función.

# Variables locales y globales
## Variables locales
Las variables locales se definen dentro de una función y pueden accederse únicamente dentro de la misma. Se crean cuando se llama a la función y se destruyen cuando esta termina de ejecutarse. Las variables locales solo se pueden acceder dentro del alcance de la función en la que están definidas. Esto significa que si una variable se define dentro de una función, es imposible acceder a ella fuera de esa función.
## Variables globales
Las variables globales se definen fuera de cualquier función y se puede acceder a ellas desde cualquier parte del código, incluso dentro de las funciones. Sin embargo, ten cuidado al utilizar variables globales, ya que pueden dificultar la comprensión y el mantenimiento del código.
## Combinar variables locales y globales
El uso de variables locales dentro de las funciones, en lugar de globales, garantiza la encapsulación y minimiza los efectos secundarios no deseados. También mejora la legibilidad del código y simplifica las pruebas y la depuración.

Sin embargo, si necesitas modificar una variable global dentro de una función, utiliza la palabra clave `global`:

````python
# Diccionario global que representa los niveles de inventario actuales
inventory = {
    'manzana': 50,
    'plátano': 30,
    'naranja': 20
}

# Muestra el inventario inicial
print("Inventario inicial:", inventory)

# Función para agregar stock
def add_stock(product, quantity):
    """
    Agrega un producto al inventario o actualiza la cantidad si ya existe.
    
    Parámetros:
    - product (str): Nombre del producto.
    - quantity (int): Cantidad a agregar.
    """
		### AQUÍ VA TU CÓDIGO ###
    global inventory # Accedemos a la variable global
    if product in inventory: # No usamos operadores lógicos o de comparación porque es un diccionario. A diferencia de una lista de diccionarios donde sí tendremos que hacerlo.
        inventory[product] += quantity
        print(f'Se agrega(n) {quantity} {product}(s). Nuevo inventario: {inventory[product]}')
    else:
        inventory[product] = quantity
        print(f'Se agrega el nuevo producto {product} con {quantity} unidades. Nuevo inventario: {inventory[product]}')


# Realiza operaciones de stock
add_stock('manzana', 20)   # Agregar 20 manzanas
add_stock('kiwi', 30)      # Agregar 30 kiwis
````

````python
# Diccionario global que representa los niveles de inventario actuales
inventory = {
    'manzana': 50,
    'plátano': 30,
    'naranja': 20
}

# Muestra el inventario inicial
print("Inventario inicial:", inventory)


# Función para quitar stock
def remove_stock(product, quantity):
	### AQUÍ VA TU CÓDIGO ###
    global inventory
    if product in inventory:
        if inventory[product] >= quantity:
            inventory[product] -= quantity
            print(f'Se quita(n) {quantity} {product}(s). Nuevo inventario: {inventory[product]}')
        else:
            print(f'Stock insuficiente de {product}. No se puede(n) quitar {quantity} {product}(s).')
    else:
        print(f'El producto {product} no existe en el inventario. No se puede quitar el stock.')

# Realiza operaciones de stock
remove_stock('plátano', 10)    # Quitar 20 plátanos
remove_stock('naranja', 25)    # Intentar quitar 25 naranjas
remove_stock('uva', 50)        # Quitar 50 uvas
````
