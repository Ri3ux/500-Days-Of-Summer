# Pandas como librería
Pandas es la librería de Python más utilizada para procesar datos. Tiene varias funcionalidades, entre ellas:

- Recuperar datos de distintas fuentes y fusionarlos;
- Solucionar problemas con los datos, por ejemplo, eliminar datos duplicados o completar valores ausentes;
- Agregar, eliminar o modificar datos de la tabla;
- Agrupar datos y calcular estadísticas para grupos.

Empecemos por conocer pandas mediante su objeto básico, que es el DataFrame.

## Estructura del DataFrame
Un DataFrame es una estructura de datos bidimensional, que es esencialmente una tabla donde cada elemento tiene dos coordenadas: una fila y una columna. Se accede a las filas por los índices y a las columnas por sus nombres.

### Atributos de un DataFrame
| Atributo | Descripción |
| --- | --- |
| `dtypes` | Permite visualizar los tipos de datos de los valores de cada columna |
| `columns` | Devuelve los nombres de las columnas del DataFrame |
| `shape` | Indica el número de filas y columnas de una tabla |

### Métodos de un DataFrame
| Método | Descripción |
| --- | --- |
| `info()` | Se utiliza para acceder a información general de una tabla (tipos de datos, valores nulos, memoria, etc.) |
| `head()` | Muestra las primeras filas de una tabla. Podemos definir la cantidad como argumento |
| `tail()` | Muestra las últimas filas de una tabla. Podemos definir la cantidad como argumento |
| `loc[]` | Permite acceder a los valores de una tabla en un DataFrame mediante fila, columna, usando **etiquetas** (indexador, se usa con corchetes) |
| `iloc[]` | Permite acceder a los valores de una tabla en un DataFrame mediante fila, columna, usando **posiciones** numéricas (indexador, se usa con corchetes) |
| `mean()` | Calcula el promedio de una columna |
| `count()` | Recuenta el número de filas |
| `sum()` | Suma todos los valores de una columna |
| `isnull().sum()` | Cuenta el número de valores nulos en cada columna |
| `notnull().sum()` | Cuenta el número de valores no nulos en cada columna |
| `isna()` | Entrega las filas que tienen valores faltantes `(NaN)` |
| `isin()` | Filtra los valores según la(s) categoría(s) que le pasamos como argumento(s) |
| `groupby()` | Agrupa los datos según la(s) categoría(s) que le pasamos como argumento(s) |
| `agg()` | Calcula diferentes métricas según las pasemos como argumento(s) |
| `describe()` | Entrega estadísticas generales de las columnas tipo `int64` o `float64`. Con el argumento `include=` podemos indicarle el tipo de objeto que queremos incluir |
| `sample()` | Entrega una muestra aleatoria de las filas. Podemos definir la cantidad como argumento |
| `rename()` | Cambia el nombre de las columnas o del índice. Se le pasa un diccionario como argumento, por ejemplo `columns={'nombre_viejo': 'nombre_nuevo'}` |
| `replace()` | Reemplaza valores específicos dentro de un DataFrame o Serie por otros valores. Se le puede pasar un valor a buscar y un valor de reemplazo, por ejemplo `df.replace('viejo', 'nuevo')`, o un diccionario para reemplazar varios valores a la vez, por ejemplo `df.replace({'viejo1': 'nuevo1', 'viejo2': 'nuevo2'})`. También se puede aplicar solo a una columna específica, por ejemplo `df['columna'] = df['columna'].replace({'viejo': 'nuevo'})`. Se puede usar `regex=True` para que los valores buscados se interpreten como expresiones regulares. A diferencia de `rename()`, que cambia nombres de columnas o índices, `replace()` cambia los **valores** dentro de las celdas del DataFrame. Por defecto no modifica el DataFrame original, sino que devuelve uno nuevo; para modificarlo directamente hay que reasignarlo (`df = df.replace(...)`) o usar `inplace=True` |
| `fillna()` | Rellena los valores faltantes `(NaN)` con el valor que le pasamos como argumento. Se le puede pasar como argumento posicional o con el parámetro explícito `value=` (por ejemplo `df.fillna(0)` o `df.fillna(value=0)`, ambos son equivalentes). También se puede pasar un diccionario para rellenar cada columna con un valor distinto (por ejemplo `df.fillna({'edad': 0, 'nombre': 'desconocido'})`). Al igual que `rename()`, por defecto `fillna()` no modifica el DataFrame original; devuelve uno nuevo. Para modificarlo directamente necesitas reasignarlo (`df = df.fillna(...)`) o usar `inplace=True` |
| `dropna()` | Elimina las filas (o columnas) que contienen valores faltantes `(NaN)`. Por defecto elimina filas (`axis=0`); para eliminar columnas se usa `axis=1` o `axis='columns'`. Se puede usar `subset=['columna1', 'columna2']` para revisar solo columnas específicas. Por defecto no modifica el DataFrame original, sino que devuelve uno nuevo; para modificarlo directamente hay que reasignarlo (`df = df.dropna()`) o usar `inplace=True` |
| `drop()` | Elimina filas o columnas específicas de un DataFrame. Para eliminar columnas se usa `columns=['col1', 'col2']` o `axis=1` junto con `labels=['col1', 'col2']`; para eliminar filas se usa `index=[...]` o `axis=0` con `labels=[...]`. Por defecto no modifica el DataFrame original, sino que devuelve uno nuevo; para modificarlo directamente hay que reasignarlo (`df = df.drop(...)`) o usar `inplace=True` |
| `duplicated()` | Devuelve una Serie booleana que indica si cada fila es duplicada (`True`) o no (`False`) respecto a las filas anteriores. Por defecto considera duplicada una fila solo si todas sus columnas coinciden con otra fila previa. Se puede usar `subset=['columna1', 'columna2']` para revisar duplicados solo en columnas específicas. El argumento `keep` controla qué ocurrencia se marca como no duplicada: `keep='first'` (por defecto, marca la primera aparición como no duplicada), `keep='last'` (marca la última) o `keep=False` (marca todas las ocurrencias como duplicadas) |
| `drop_duplicates()` | Elimina las filas duplicadas de un DataFrame. Por defecto compara todas las columnas para determinar si una fila es duplicada. Se puede usar `subset=['columna1', 'columna2']` para considerar duplicados solo según columnas específicas. El argumento `keep` controla cuál ocurrencia se conserva: `keep='first'` (por defecto, conserva la primera), `keep='last'` (conserva la última) o `keep=False` (elimina todas las ocurrencias duplicadas). Por defecto no modifica el DataFrame original, sino que devuelve uno nuevo; para modificarlo directamente hay que reasignarlo (`df = df.drop_duplicates()`) o usar `inplace=True` |
| `reset_index()` | Reinicia el índice de un DataFrame, asignando una nueva secuencia numérica consecutiva (0, 1, 2, ...). Es útil después de filtrar filas o eliminar duplicados, ya que el índice original queda con "huecos". Por defecto conserva el índice anterior como una nueva columna llamada `index`; para descartarlo se usa `drop=True`. Por defecto no modifica el DataFrame original, sino que devuelve uno nuevo; para modificarlo directamente hay que reasignarlo (`df = df.reset_index()`) o usar `inplace=True` |
| `unique()` | Devuelve un arreglo (`array`) con los valores únicos de una columna (Serie), sin repetir, y en el orden en que aparecen por primera vez. Solo se puede aplicar sobre una Serie, no sobre un DataFrame completo, por ejemplo `df['columna'].unique()` |
| `nunique()` | Devuelve el número (conteo) de valores únicos distintos que tiene una columna o un DataFrame. Por defecto no cuenta los valores nulos `(NaN)` dentro del conteo; para incluirlos se usa `dropna=False` |
| `groupby()` | Agrupa las filas de un DataFrame según la(s) categoría(s) (columna o columnas) que se le pasan como argumento, permitiendo aplicar luego una función de agregación a cada grupo (por ejemplo `df.groupby('columna')['otra_columna'].mean()`). Se pueden agrupar por varias columnas a la vez pasando una lista, por ejemplo `df.groupby(['columna1', 'columna2'])`. El resultado por sí solo es un objeto `GroupBy`, no una tabla visible; necesita combinarse con un método de agregación como `mean()`, `sum()`, `count()` o `agg()` para mostrar resultados. Por defecto ordena los grupos según la categoría (`sort=True`) |
| `sort_values()` | Ordena las filas de un DataFrame (o una Serie) según los valores de una o más columnas, que se pasan como argumento en `by=` (por ejemplo `df.sort_values(by='columna')` o `df.sort_values(by=['columna1', 'columna2'])`). Por defecto ordena de forma ascendente; para ordenar de forma descendente se usa `ascending=False` (o una lista de booleanos si se ordena por varias columnas con direcciones distintas). Por defecto no modifica el DataFrame original, sino que devuelve uno nuevo; para modificarlo directamente hay que reasignarlo (`df = df.sort_values(...)`) o usar `inplace=True` |
| `value_counts()` | Cuenta cuántas veces aparece cada valor único en una columna (Serie) y devuelve el resultado ordenado de mayor a menor frecuencia por defecto. Solo se puede aplicar sobre una Serie, no sobre un DataFrame completo, por ejemplo `df['columna'].value_counts()`. Por defecto no cuenta los valores nulos `(NaN)`; para incluirlos se usa `dropna=False`. Para obtener las frecuencias como proporciones (porcentajes) en lugar de conteos, se usa `normalize=True`. Para ordenar de forma ascendente en lugar de descendente, se usa `ascending=True` |
| `to_csv()` | Método de un DataFrame que exporta sus datos a un archivo `.csv`, por ejemplo `df.to_csv('/datasets/salida.csv')`. Parámetros comunes: `index=` para indicar si se debe incluir el índice del DataFrame como columna en el archivo (por defecto `True`; se suele usar `index=False` para no guardarlo); `sep=` para indicar el separador de columnas (por defecto `,`); `columns=` para exportar solo columnas específicas; `header=` para indicar si se incluyen los nombres de las columnas (por defecto `True`), o para renombrarlas al exportar pasando una lista; `na_rep=` para indicar cómo representar los valores `NaN` en el archivo (por defecto los deja como cadena vacía); `encoding=` para indicar la codificación del archivo (por ejemplo `'utf-8'`); `mode=` para indicar si se sobrescribe (`'w'`, por defecto) o se agrega al final (`'a'`) del archivo |

### Funciones de Pandas
| Función | Descripción |
| --- | --- |
| `read_csv()` | Función de pandas (no un método de DataFrame, se llama como `pd.read_csv()`) que lee un archivo `.csv` y lo carga como un DataFrame. El primer argumento es la ruta o URL del archivo, por ejemplo `pd.read_csv('/datasets/archivo.csv')`. | `read_excel()` | Función de pandas (no un método de DataFrame, se llama como `pd.read_excel()`) que lee un archivo `.xlsx` o `.xls` y lo carga como un DataFrame. Requiere tener instalada una librería adicional como `openpyxl`. El primer argumento es la ruta del archivo, por ejemplo `pd.read_excel('/datasets/archivo.xlsx')`. Parámetros comunes: `sheet_name=` para indicar qué hoja leer (por nombre, número, o `None` para leer todas las hojas y obtener un diccionario de DataFrames); `header=` para indicar qué fila usar como encabezado; `names=` para asignar nombres de columnas manualmente; `usecols=` para leer solo columnas específicas; `dtype=` para forzar el tipo de dato de columnas; `na_values=` para indicar valores adicionales que deben tratarse como `NaN`; `skiprows=` para saltar un número de filas al inicio del archivo; `nrows=` para leer solo una cantidad limitada de filas |

## Notación abreviada para la indexación
| Tipo | Notación completa | Notación abreviada |
|------|-------------------|---------------------|
| Una celda | `.loc[7, 'genre']` | - |
| Una columna | `.loc[:, 'genre']` | `df['genre']` |
| Columnas múltiples | `.loc[:, ['genre', 'Artist']]` | `df[['genre', 'Artist']]` |
| Múltiples columnas consecutivas (slice) | `.loc[:, 'total play':'genre']` | - |
| Una fila | `.loc[1]` | - |
| Todas las filas, empezando por la fila dada | `.loc[1:]` | `df[1:]` |
| Todas las filas, hasta la fila dada | `.loc[:3]` incluyendo 3 | `df[:3]` sin incluir 3 |
| Múltiples filas consecutivas (slice) | `.loc[2:5]` incluyendo 5 | `df[2:5]` sin incluir 5 |

## Indexación lógica booleana
Esta forma de indexación lógica es muy común. Puedes utilizar expresiones lógicas con cualquier operación lógica:

- Para comprobar la igualdad o la desigualdad, respectivamente, utiliza == y !=.
- Para comprobar si un valor es mayor o menor, utiliza > o <, respectivamente.
- Utiliza >= para comprobar si un valor es mayor o igual que y <= para comprobar si un valor es menor o igual que.

Todas las expresiones lógicas devolverán los valores lógicos `True` o `False`.

```python
import pandas as pd

df = pd.read_csv('/datasets/music_log_chpt_11.csv')

# selecciona las filas en las que el género es el jazz y el tiempo total de reproducción oscila entre 80 y 130
df = df[df['total play'] >= 80]
df = df[df['total play'] <= 130]
df = df[df['genre'] == 'jazz']

print(df)
```

# Preprocesamiento de datos
## Renombrar columnas
Creemos una tabla que contenga las distancias entre la Tierra y varios cuerpos celestes. Crearemos un DataFrame a partir de estos datos.

```python
import pandas as pd

# Las medidas se almacenan en una lista de listas
measurements = [['Sun', 146, 152],
                ['Moon', 0.36, 0.41], 
                ['Mercury', 82, 217], 
                ['Venus', 38, 261],
                ['Mars', 56, 401],
                ['Jupiter', 588, 968],
                ['Saturn', 1195, 1660],
                ['Uranus', 2750, 3150],
                ['Neptune', 4300, 4700],
                ['Halley\'s comet', 6, 5400]]

# Los nombres de las columnas se almacenan en la variable header
header = ['Celestial bodies ','MIN', 'MAX']

# Guardammos el DataFrame en la variable celestial

celestial = pd.DataFrame(data = measurements, columns = header)
print(celestial)
```

Debemos corregir estos problemas, así que cambiemos el nombre de las columnas:

- Vamos a cambiar `'Celestial bodies '` a `'celestial_bodies'`, para respetar el estilo `snake_case`, y de esta manera resolvemos los problemas del nombre de esta columna.
- Remplazaremos `'MIN'` y `'MAX'` por `'min_distance'` y `'max_distance'`, y de esa manera también respetaremos el estilo snake_case y agregaremos significado al nombre de las columnas, lo que hace que su contenido se entienda más fácilmente.

```python
# Declara un diccionario con el nombre anterior de la columna como claves
# y los nombres nuevos de la columna como los valores
columns_new ={
    "Celestial bodies ": "celestial_bodies",
    "MIN": "min_distance",
    "MAX": "max_distance",
    }

# Llama al método rename y pasa el diccionario como un argumento al parámetro columns
celestial = celestial.rename(columns = columns_new)
print(celestial)
```

Hay una forma más elegante de renombrar columnas que no requiere reasignación como hicimos anteriormente. Solo necesitamos especificar el parámetro `inplace=` y establecerlo en `True`.

```python
# Declara un diccionario con el nombre anterior de la columna como claves
# y los nombres nuevos de la columna como los valores
columns_new ={
    "Celestial bodies ": "celestial_bodies",
    "MIN": "min_distance",
    "MAX": "max_distance",
    }

celestial.rename(columns = columns_new, inplace = True)
print(celestial.shape)
```

### Forma automatizada de renombrar las columnas
A veces, la cantidad de columnas en un dataset puede ser grande, por lo que es poco práctico asignar manualmente nuevos valores a los nombres de las columnas. Y de hecho, podría ser difícil ver los problemas en un nombre de columna. En estos casos, los bucles y los métodos de string pueden ser muy útiles. 

```python
new_column_names = []

for old_name in celestial.columns:
    # Primero, elimina los espacios al principio y al final
    name_stripped = old_name.strip()
    # Luego, pon todas las letras en minúsculas
    name_lowered = name_stripped.lower()
    # Por último, reemplaza los espacios entre palabras por guiones bajos
    name_no_spaces = name_lowered.replace(' ', '_')
    # Agrega el nuevo nombre a la lista de nuevos nombres de columna
    new_column_names.append(name_no_spaces)

celestial.columns = new_column_names
print(celestial.columns)
```

## Procesar valores ausentes
El procesamiento de valores ausentes es fundamental. Los algoritmos de aprendizaje automático no funcionan bien con los datos que contienen valores ausentes. Además, los algoritmos estadísticos que pueden ayudar a las empresas a tomar decisiones informadas también sufren problemas con la información incompleta. Como regla general, siempre debemos abordar los valores ausentes antes de seguir adelante.

Los valores ausentes pueden aparecer de varias maneras, pero aquí hay dos de las más comunes:

1. Formas esperadas: `None` o `NaN`. Estos son dos valores especiales que marcan los valores ausentes:
- `None` es una instancia del objeto `NoneType`. En términos más sencillos, se utiliza para representar un valor nulo o la ausencia de cualquier valor.
- `NaN` significa "no es un número" (en inglés, "Not A Number") y se utiliza para representar valores ausentes en los datos. Se trata como un decimal, específicamente como un tipo `float`. `NaN` representa cualquier valor que no está definido o no se puede representar.
2. Formas inesperadas. Las personas pueden optar por marcar los valores ausentes de formas alternativas, tales como `0`, '`?`', '`NN`' o '`n/a`'. Lo ideal es que también proporcionen documentación para los datos. Si no lo hacen, tendrás que hacer una conjetura sobre el significado de los valores ausentes.

Si encuentras con frecuencia un carácter especial en una tabla sin ninguna explicación, es probable que represente valores ausentes. Y los tratarás del mismo modo en que tratas `None` o `NaN`, pero observa que Python solo reconoce a `None` y `NaN` como valores ausentes, así que tendrás que identificarlos manualmente.

```python
import pandas as pd

df = pd.read_csv('Datasets/blockbuster_movies.csv')

print(df.isna().sum()) # Cuenta los valores True y devuelve la suma total
```

```python
import pandas as pd

df = pd.read_csv('Datasets/blockbuster_movies.csv')

df['budget_millions'] = df['budget_millions'].fillna(0)
df['box_office_millions'] = df['box_office_millions'].fillna(0)
print(df)
```

### Forma automatizada de procesar valores ausentes
A veces, la cantidad de columnas en un dataset puede ser grande, por lo que es poco práctico procesar manualmente los valores ausentes de cada columna. En estos casos, los bucles y los métodos de string pueden ser muy útiles. 

```python
import pandas as pd

df = pd.read_csv('Datasets/blockbuster_movies.csv')
columns_to_process = ['budget_millions', 'box_office_millions', 'rating'] # Definimos una lista con los nombres de las columnas a procesar

for column in columns_to_process:
    df[column] = df[column].fillna(0) # Podríamos usar el argumento 'inplace=' para saltarnos la reasignación

print(df)
```

### Eliminar filas
Para eliminar filas con valores ausentes en un DataFrame de pandas, usa el método `dropna()`. Este método elimina las filas con al menos un valor ausente. También puedes especificar una lista de columnas en su parámetro `subset=` para que elimine filas con valores nulos solo en esas columnas.

```python
import pandas as pd

df = pd.read_csv('Datasets/blockbuster_movies.csv')

df = df.dropna(subset=['title']) # Elimina todas las filas de la lista de columnas que contienen valores ausentes

print(df)
```

Ahora eliminemos toda la columna 'notes', que consiste casi en su totalidad en valores ausentes.

Usaremos el método `dropna()` de nuevo, pero esta vez agregaremos otro argumento: `axis=`. Este argumento nos permite especificar si queremos eliminar filas o columnas. Si pasamos el string `'columns'` a `axis=`, se eliminarán las columnas que tengan valores ausentes. Dado que `notes` es la única columna que contiene valores ausentes, podemos usar esta opción de forma segura para eliminarla.

```python
import pandas as pd

df = pd.read_csv('Datasets/blockbuster_movies.csv')

df = df.dropna(axis='columns') # Elimina todas las columnas que tienen valores ausentes. Podemos usar el argumento 'inplace' para no hacer una reasignación

print(df)
```

Conviene señalar que en los DataFrames, a diferencia de las listas de Python, los índices siguen siendo los mismos cuando se eliminan las filas.

Cuando eliminas filas, a menudo también es importante actualizar el índice. Para hacerlo, llama al método `reset_index()`. Esto creará un nuevo DataFrame en el que:

- Los índices del DataFrame original se ubicarán en una nueva columna llamada `index`.
- Los nuevos índices se establecerán en orden para todas las filas en el DataFrame.

Así es como restablecemos los índices:

```python
import pandas as pd
df = pd.read_csv('Datasets/music_listening_data.csv')
# Como resultado, obtuvimos la enumeración correcta para nuestras filas y también para la columna 'index', que básicamente almacena los índices anteriores. Por lo general, queremos eliminar esta columna 'index'.
# Para ello, necesitamos establecer el parámetro drop= en True:
df = df.drop_duplicates().reset_index(drop=True)

print(df)
```

### Automatización con funciones personalizadas
Para evitar repetir el mismo código varias veces, los profesionales de los datos suelen escribir sus propias funciones. Vamos a crear una función que tome cuatro argumentos:

- el DataFrame;
- el nombre de la columna donde queremos realizar el reemplazo;
- una lista de valores incorrectos;
- el valor correcto.

La función reemplazará todos los valores incorrectos por el correcto en la columna seleccionada.

```python
import pandas as pd

rating = ['date', 'name', 'points']
players = [
        ['2018.01.01',  'Rafael Nadal', 10645],
        ['2018.01.08',  'Rafael Nadal', 10600],
        ['2018.01.29',  'Rafael Nadal', 9760],
        ['2018.02.19',  'Roger Federer', 10105], 
        ['2018.03.05',  'Roger Federer', 10060],
        ['2018.03.19',  'Roger Federerr', 9660],
        ['2018.04.02',  'Rafael Nadal Parera', 8770],
        ['2018.06.18',  'Roger Fedrer', 8920],
        ['2018.06.25',  'Rafael Nadal Parera', 8770],
        ['2018.07.16',  'Rafael Nadal Parera', 9310],
        ['2018.08.13',  'Rafael Nadal Parera', 10220],
        ['2018.08.20',  'Rafael Nadal Parera', 10040],
        ['2018.09.10',  'Rafael Nadal Parera', 8760],
        ['2018.10.08',  'Rafael Nadal Parera', 8260],
        ['2018.10.15',  'Rafael Nadal Parera', 7660],
        ['2018.11.05',  'Novak Djokovic', 8045],
        ['2018.11.19',  'Novak Djokovic', 9045]
]

tennis = pd.DataFrame(data=players, columns=rating)

def replace_wrong_values(dataframe, column, wrong_values, correct_value): # Pasa una lista de valores incorrectos y un string con el valor correcto en la entrada de la función
    for wrong_value in wrong_values: # Un bucle sobre nombres mal escritos
        dataframe[column] = dataframe[column].replace(wrong_value, correct_value) # Llama a replace() para cada nombre incorrecto
    return dataframe # Devuelve el DataFrame modificado

duplicates = ['Roger Federerr', 'Roger Fedrer'] # Una lista de nombres mal escritos
player_name = 'Roger Federer' # El nombre correcto

tennis = replace_wrong_values(tennis, 'name', duplicates, player_name) # Llama a la función, replace() se llamará dos veces
print(tennis) # El nuevo DataFrame sin duplicados
```

### Agrupación en Pandas
En Pandas, agrupamos los datos utilizando el método `groupby()`, que hace lo siguiente:

- Toma el nombre de una columna en la que se deben agrupar los datos como argumento. Este parámetro se llama `by=`. En nuestro caso, vamos a agrupar los datos por año de descubrimiento.
- Devuelve un objeto de un tipo especial: `DataFrameGroupBy`. Son datos agrupados. Si les aplicas un método de pandas, se convertirán en una nueva estructura de datos.

```python
import pandas as pd

exoplanet = pd.read_csv('Datasets/exoplanets.csv')

print(exoplanet.groupby(by='discovered'))
print()
print(exoplanet.groupby(by='discovered').count())
```

## Ordenar datos
El método `sort_values()` en pandas es una poderosa herramienta para ordenar valores. Puede ordenar la tabla completa o grupos de filas dentro de ella.

Este método se aplica a un DataFrame y tiene dos parámetros:

- `by=`: el nombre o nombres de la columna cuyos valores se usan para ordenar las filas del DataFrame.
- `ascending=`: indica el orden al realizar el ordenamiento. Su valor predeterminado es `True`. Para ordenar los datos en orden descendente, establece este parámetro en `False`.

```python
import pandas as pd
exoplanet = pd.read_csv('Datasets/exoplanets.csv')

print(exoplanet.sort_values(by='radius').head(10))
```

Podemos extraer todos los exoplanetas con un radio menor al de la Tierra `(radius = 1)` utilizando la indexación lógica:

```python
import pandas as pd
exoplanet = pd.read_csv('Datasets/exoplanets.csv')

print(exoplanet[exoplanet['radius'] < 1].sort_values(by='radius'))
```

Ahora combninaremos dos condiciones para encontrar estos planetas:

- Planetas descubiertos en 2011.
- Planetas más pequeños que la Tierra.

```python
import pandas as pd
exoplanet = pd.read_csv('Datasets/exoplanets.csv')

exo_small_2011 = exoplanet[exoplanet['radius'] < 1]
exo_small_2011 = exo_small_2011[exo_small_2011['discovered'] == 2011]

print(exo_small_2011.sort_values(by='radius', ascending=False))
```

El método `sort_values()` devuelve un nuevo objeto en vez de modificarlo localmente. Por lo tanto, si deseas seguir trabajando con el DataFrame ordenado, deberás almacenar el resultado en una variable. Puedes guardarla nuevamente en la misma:

```python
exo_small_2011 = exo_small_2011.sort_values(by='radius', ascending=False)
print(exo_small_2011)
```

# Rellenar los valores ausentes cuantitativos
Para decidir si la media o la mediana es un valor más representativo, podemos seguir estos pasos:

1. Determina si los datos tienen valores atípicos significativos.
2. Si no hay valores atípicos significativos, calcula la media utilizando el método `mean()`.
3. Si tus datos tienen valores atípicos significativos, calcula la mediana utilizando el método `median()`.
4. Reemplaza los valores ausentes con la media o la mediana utilizando el método `fillna()`.

También vale la pena señalar que a veces no necesitamos rellenar los valores ausentes en absoluto.  Por ejemplo, si solo falta una pequeña parte de tus datos, y los datos ausentes son aleatorios, podría ser buena idea dejar los valores como `NaN`, en cuyo caso simplemente no se incluirían en ningún cálculo numérico.
